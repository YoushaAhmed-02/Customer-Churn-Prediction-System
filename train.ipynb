## Data Cleaning

```python
df.drop("customerID", axis=1, inplace=True)
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ", np.nan)

df["TotalCharges"] = pd.to_numeric(
    df["TotalCharges"],
    errors="coerce"
)

df["TotalCharges"] = df["TotalCharges"].fillna(
    df["TotalCharges"].median()
)
```

```python
for col in df.columns:
    blanks = (df[col].astype(str).str.strip() == "").sum()
    print(f"{col}: {blanks}")
```

## Feature Engineering

```python
df["Churn"] = df["Churn"].map({
    "Yes": 1,
    "No": 0
})
```

```python
selected_features = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges",
    "Contract",
    "InternetService",
    "PaymentMethod"
]

X = df[selected_features].copy()

y = df["Churn"]
```

```python
encoders = {}

categorical_cols = [
    "Contract",
    "InternetService",
    "PaymentMethod"
]
for col in categorical_cols:

    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    encoders[col] = le
```

```python
print(X.dtypes)
print(y.isnull().sum())
```

## Train & Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

## Model Training

```python
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)
model.fit(X_train, y_train)
```

## Prediction

```python
y_pred = model.predict(X_test)
```

## Model Evaluation

```python
accuracy = accuracy_score(
    y_test,
    y_pred
)
print(f"Model Accuracy: {accuracy*100:.2f}%")
```

```python
print(classification_report(
    y_test,
    y_pred
))
```

```python
print(confusion_matrix(
    y_test,
    y_pred
))
```

```python
joblib.dump(
    model,
    "models/model.pkl"
)

joblib.dump(
    encoders,
    "models/encoders.pkl"
)
```
